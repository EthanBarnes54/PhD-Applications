```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

Angle_Range = np.arange(0, 2*np.pi, 0.05)

Coefficent_1 = []
Coefficent_2 = [] 
Coefficent_3 = [] 
Coefficent_4 = []
Coefficent_5 = []

for i in range(0,len(Angle_Range)):

    Coefficent_1.append((np.random.random()*1/2 - 1/4)*0.05)
    Coefficent_2.append((np.random.random()*1/2 - 1/4) *0.05)
    Coefficent_3.append((np.random.random()*1/2 - 1/4) *0.05)
    Coefficent_4.append((np.random.random()*1/2 - 1/4) *0.05)
    Coefficent_5.append((np.random.random()*1/2 - 1/4) *0.05)

Constant_Phase_Shift = 0.5

Gussian_Standard_Deviation = 4
Stochastic_Phase_Shifts= []

for i in range(0,len(Angle_Range)):

    Stochastic_Phase_Shifts.append(2*np.pi*np.random.random()*(1/(Gussian_Standard_Deviation*np.sqrt(2*np.pi)))*np.exp((-Angle_Range[i]-np.pi**2)/(2*Gussian_Standard_Deviation**2)))

y_A = []
y_B = []

for i in range(0,len(Angle_Range)):

    y_A.append(Coefficent_1[i] + (1+Coefficent_2[i])*np.sin(Constant_Phase_Shift+Stochastic_Phase_Shifts[i]+Coefficent_3[i]))
    y_B.append(Coefficent_4[i] + (1+Coefficent_5[i])*np.sin(Stochastic_Phase_Shifts[i]))

    y_A.append(np.abs(Coefficent_1[i] + (1+Coefficent_2[i])*np.sin(Constant_Phase_Shift+Stochastic_Phase_Shifts[i]+Coefficent_3[i]) - np.pi))
    y_B.append( np.abs(Coefficent_4[i] + (1+Coefficent_5[i])*np.sin(Stochastic_Phase_Shifts[i]) - np.pi))

plt.figure()
plt.scatter(y_A, y_B)
plt.title("Manufactured Data Sample for Inertial Navigation System")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.show()

plt.figure()
plt.plot(Angle_Range, Stochastic_Phase_Shifts)
plt.title("Stochastic Variations in Phase Measurements")
plt.xlabel("Phase Measurement (Rad)")
plt.ylabel("Stochastic Variation to Phase (Rad)")
plt.show()
```
